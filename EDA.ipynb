# Human Activity Recognition

This project is to build a model that predicts the human activities from tri-axial smartphone accelerometer data, which have been numbered with integers. 
We use the following encoding: 1 = standing, 2 = walking, 3 = stairs down, 4 = stairs up.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn as sns
sns.set()
```

### Load and View Input Data

```python
train_ts = pd.read_csv("train_time_series.csv", index_col=0)
train_labels = pd.read_csv("train_labels.csv", index_col=0)
print("Shape of time series signals data:", train_ts.shape)
print("Shape of labels data:", train_labels.shape)
```

##### Plotting # of data points per activity

```python
plt.title('No of Datapoints per Activity', fontsize=15)
sns.countplot(train_labels.label)
plt.show()
```

Our data isn't well balanced

##### Time-series Data Visualization

```python
plt.figure(figsize=(25,10))
plt.plot(train_ts['timestamp'], train_ts['x'], label='x-component')
plt.plot(train_ts['timestamp'], train_ts['y'], label='y-component')
plt.plot(train_ts['timestamp'], train_ts['z'], label='z-component')
plt.xlabel('timestamp', fontsize=15)
plt.ylabel('acceleration',fontsize=15)
plt.legend(loc='upper left')
plt.title('Time-series Data Visualization per component', fontsize=20)
```

```python
def magnitude(activity):
    x2 = activity['x'] * activity['x']
    y2 = activity['y'] * activity['y']
    z2 = activity['z'] * activity['z']
    m2 = x2 + y2 + z2
    m = m2.apply(lambda x: math.sqrt(x))
    return m

train_ts['m'] = magnitude(train_ts)
train_ts.head()
```

#### Data Visulization by Activity

We will merge magnitudes of acceleration to plot activities' behaviours

```python
act_behaviours= pd.merge(train_labels,train_ts, on='timestamp')


facetgrid = sns.FacetGrid(act_behaviours, hue='label', size=6,aspect=2)
facetgrid.map(sns.distplot,'m', hist=False)
plt.legend(loc='upper left')
plt.xlabel('Magnitude of acceleration', fontsize=15)
plt.title('Data Visualization per activity', fontsize=17)
plt.annotate("Stationary Activities", xy=(1.05,8), xytext=(1.8,9), size=15,\
            va='center', ha='left',\
            arrowprops=dict(arrowstyle="simple",connectionstyle="arc3,rad=0.1"))

plt.annotate("Moving Activities", xy=(1.5,0.7), xytext=(1.7, 2), size=15,\
            va='center', ha='left',\
            arrowprops=dict(arrowstyle="simple",connectionstyle="arc3,rad=0.1"))
```

For plotting purposes taking datapoints of each activity to a different dataframe

```python
df1 = act_behaviours[act_behaviours['label']==1]
df2 = act_behaviours[act_behaviours['label']==2]
df3 = act_behaviours[act_behaviours['label']==3]
df4 = act_behaviours[act_behaviours['label']==4]

plt.figure(figsize=(14,7))
plt.subplot(2,2,1)
plt.title('Stationary Activity')
sns.distplot(df1 ['m'],color = 'r',hist = False, label = 'Standing')
plt.legend(loc='center right')
plt.xlabel('Magnitude of acceleration')

plt.subplot(2,2,2)
plt.title('Moving Activities')
sns.distplot(df2['m'],color = 'red',hist = False, label = 'Walking')
sns.distplot(df3['m'],color = 'blue',hist = False,label = 'Stairs down')
sns.distplot(df4['m'],color = 'green',hist = False, label = 'Stairs Up')
plt.legend(loc='center right')
plt.xlabel('Magnitude of acceleration')


plt.tight_layout()
plt.show()
```
